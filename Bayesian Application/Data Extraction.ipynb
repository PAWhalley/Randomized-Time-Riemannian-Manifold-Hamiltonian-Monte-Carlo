```python
#print(data_matrix[40:80,-1])
n = 2000
print(data_matrix[:10,0])
```

```python
mean = np.zeros(120)
for i in range(n):
    mean += data_matrix[:,i]/n
```

```python
#print(mean)
```

```python
S = np.zeros((120,120))
for i in range(n):
    x = data_matrix[:,i] - mean
    S += np.outer(x,x)/n
```

```python
plt.matshow(S,cmap='jet')
plt.colorbar()
plt.show()
```

```python
log_S = np.zeros((120,120))
for i in range(120):
    for j in range(120):
        log_S[i,j] = np.log(abs(S[i,j]))
```

```python
plt.matshow(log_S,cmap='jet')
plt.colorbar()
plt.show()
```

```python
S_inv = np.linalg.inv(S)
```

```python
plt.matshow(S_inv,cmap='jet')
plt.colorbar()
plt.show()
```

```python
log_S_inv = np.zeros((120,120))
for i in range(120):
    for j in range(120):
        log_S_inv[i,j] = np.log(abs(S_inv[i,j]))
```

```python
plt.matshow(log_S_inv,cmap='jet')
plt.colorbar()
plt.show()
```
